# Indexando o DataFrame

## Usando o operador [ ]
 * O nome de uma coluna
 * Uma lista com nomes de colunas

```python
df[ 'director_name' ]
```

```python
df[ ['director_name', 'color' ] ]
```

## Usando .loc
Sintaxe: .loc[ < expressao pra linhas >, <expressão para colunas>]
A expressão das colunas é opcional, se não passar ele retorna todas as colunas.

As expressões aceitas pelo .loc são sempre os índices.
As possíveis expressões são:

* o valor do índice
* uma lista de índices
* um slice

```python
df.loc[ 0, 'director_name' ]
```

```python
df.loc[ 0 ]
```

```python
df.loc[ 0:3, ['color', 'director_name' ] ]
```

```python
df.loc[ 0:3, 'color':'director_name'  ]
```

```python
df.loc[ 0:3, 'color':'duration'  ]
```

# Usando Filtros (indexação booleana)

```python
df['director_name']
```

```python
df['director_name'] == 'James Cameron'
```

```python
df[ df['director_name'] == 'James Cameron' ]
```

### Seleção booleana usando .loc

```python
df.loc[ df['director_name'] == 'James Cameron', ['duration','num_critic_for_reviews' ] ]
```

```python
df.loc[ (df['director_name'] == 'James Cameron') & (df['duration'] < 120)  ]
```

```python
df.loc[ (df['director_name'] == 'James Cameron') | (df['duration'] < 120)  ]
```

## Indexando com .iloc

```python
df.iloc[ 0:3, [ 0, 1, 2, 3, 8, 9, 10 ] ]
```

```python
df.iloc[ 0:3, list(range(0,5)) + list(range(5,8))  ]
```

```python
df.iloc[ 0:3, 0 ]
```

```python
# .iloc não aceita receber arrays de booleanos para indexar
df.iloc[ df['director_name'] == 'James Cameron' ]
```

```python
cameron = df[ df['director_name'] == 'James Cameron' ]
```

```python
cameron
```

```python
cameron.loc[ 0:26 ]
```

```python
cameron.iloc[ 0:5 ]
```

```python
cameron.reset_index()
```
