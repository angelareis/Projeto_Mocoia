```python
X = np.linspace( -np.pi, np.pi, 256 )
C = np.cos(X)
S = np.sin(X)
```

```python
plt.plot(X, C, color='red', linewidth=3, linestyle='-.', label='Cos')
plt.legend()
plt.show()

plt.plot(X, S, label='Sin')
#plt.xlim(-1, 1)
#plt.ylim(-10, 1)
#plt.xticks( [-np.pi, -np.pi/2, 0, np.pi/2, np.pi] )
plt.yticks( [ -1, -0.5, 0, 0.5, 1.0] )
plt.title('Senoides')
plt.xlabel('X')
plt.ylabel('Valor')
plt.legend()
#plt.savefig('figura.png', dpi=300)
```

# Boxplot

Usado para detecção de outliers. Outliers são os pontos que parecem não pertencer a distribuição. A linha no meio da caixa marca a mediana e a caixa marca o intervalo que contem 50% dos dados. Os pontos acima ou abaixo dos whisker são possíveis outliers e necessitam de atenção.

```python
plt.boxplot( tips['total_bill'] );
```

```python
plt.boxplot( tips['tip'] );
```

```python
np.median(tips['total_bill'])
```

```python
plt.violinplot( tips['total_bill'] );
```

# Plotando Matrizes

O comando imshow() é usado para visualizar matrizes, associando uma cor ao valor de cada célula.

```python
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 10
x = np.linspace(-3, 3, 4*n)
y = np.linspace(-3, 3, 3*n)
X, Y = np.meshgrid(x,y)

m = f(X,Y)
```

```python
m.shape
```

```python
m
```

```python
plt.imshow( m, cmap='viridis' )
plt.colorbar()
```

# Usando subplots para criar vários axes

```python
plt.subplot( 2, 2, 1 )
plt.imshow( m, cmap='jet' )

plt.subplot( 2, 2, 2 )
plt.boxplot( tips['tip'] );

plt.subplot( 2, 2, 3)
plt.boxplot( tips['tip'] );

plt.subplot( 2, 2, 4 )
plt.hist( tips['tip'] );
```
